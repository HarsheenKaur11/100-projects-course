```python
#Step 2 (Logic: generate the blank list and print the letter if in the list)
import random
list1=["hello","apple","custard"]
ch=random.choice(list1)
display=[]
for i in ch:
    display+="_"
print(display)
guess=input("Guess a letter ").lower()
for position in range(len(ch)):
    letter=ch[position]
    if letter==guess:
        display[position]=letter
print(display)
```

```python
#Step 3 (Logic: loop check and check if the player won)
import random
list1=["hello","apple","custard"]
ch=random.choice(list1)
display=[]
for i in ch:
    display+="_"
print(display)
end_of_game=False
while not end_of_game:
    guess=input("Guess a letter ").lower()
    for position in range(len(ch)):
        letter=ch[position]
        if letter==guess:
            display[position]=letter
    print(display)
    if "_" not in display:
        end_of_game=True
```

```python
#Step 4 (Logic: Hangman figure-tracking lives of the player)
import random
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

list1=["hello","apple","custard"]
ch=random.choice(list1)
lives=6

display=[]
for i in ch:
    display+="_"
print(display)

end_of_game=False
while not end_of_game:
    guess=input("Guess a letter ").lower()
    for position in range(len(ch)):
        letter=ch[position]
        if letter==guess:
            display[position]=letter
    if guess not in ch:
        lives-=1
        if lives==0:
            end_of_game=True
            print("You lose!")
    print(display)
    if "_" not in display:
        end_of_game=True
        print("You Win!")
    print(stages[lives])
```
